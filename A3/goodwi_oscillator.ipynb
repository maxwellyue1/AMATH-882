```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#assign parameter values

a1=360
kappa1=43
k1=1
b1=1

alpha1=1
beta1=0.6

gamma1=1
delta1=0.8

xi = 1
psi = 0.7

n=6
```

```python
#set time_grid for simulation
times=np.linspace(0, 55, 1000) #generate time-grid list
```

```python
#assign initial condition.  
#state vector S=[X Y Z]
Sinit_4=[ 0,0,0,0]
```

```python
#declare right-hand-side for original model
def dSdt_modified(S,t):
    dS=np.zeros(4) #generate a list to store derivatives
    
    X=S[0]
    Y=S[1]
    Z=S[2]
    W=S[3]
    
    dS[0]=a1/(kappa1+k1*W**n) - b1*X
    dS[1]=alpha1*X - beta1*Y
    dS[2]=gamma1*Y - delta1*Z
    dS[3]=xi*Z - psi*W

    return dS
```

```python
S=odeint(dSdt_modified, Sinit_4, times) #run simulation
```

```python
#plot simulation
plt.figure() #generate figure 7.17A
plt.plot(times, S[:,0], label="mRNA (X)", linewidth=2)
plt.plot(times, S[:,1], label="enzyme (Y)", linewidth=2)
plt.plot(times, S[:,2], label="metabolite (Z)", linewidth=2)
plt.plot(times, S[:,3], label="Additional step (W)", linewidth=2)
plt.xlabel("Time")
plt.ylabel("concentration")
plt.legend()
plt.title("Modified model")

#generate figure 7.17B

plt.figure() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from jitcdde import jitcdde, y, t

# Given parameters
a = 10
k = 1
b = 0.5
n = 4

# Function to solve and plot for different delays
def solve_dde(tau, t_max=200):
    # Define the delay differential equation
    def dde_equation():
        return [ a / (k**n + y(0, t - tau)**n) - b * y(0) ]
    
    # Initialize DDE system
    dde = jitcdde(dde_equation())  # Fix: Call function to return the equation list
    dde.constant_past([1.0])  # Initial history function (constant past)
    dde.step_on_discontinuities()

    # Time range
    times = np.linspace(0, t_max, 1000)
    results = [dde.integrate(time)[0] for time in times]  # Extract solution from array

    # Plot the results
    plt.plot(times, results, label=f"τ = {tau}")

# Run simulations for different delays
plt.figure(figsize=(10,6))
for tau in [2, 0.75, 20]:  # Given values
    solve_dde(tau)

# Plot formatting
plt.xlabel("Time")
plt.ylabel("x(t)")
plt.title("Effect of Time Delay on Oscillations")
plt.legend()
plt.grid()
plt.show()
```
